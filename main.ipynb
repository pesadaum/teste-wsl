```python
import matplotlib.pyplot as plt
import numpy as np

rc = {"font.family": "serif",  # fonte serifada
      "mathtext.fontset": "cm",  # texto matemático em LaTeX
    #   "font.size": 12,  # Usar tamanho 12 ou 14
      "axes.axisbelow": True,  # grid abaixo dos dados
      "text.usetex": True,
      "axes.grid": True
      }

%matplotlib inline

plt.rcParams.update(rc)


x = np.arange(-6, 6, 0.01)
y = np.exp(-x)*np.cos(10*x)


plt.plot(x, y)
plt.plot(x, np.exp(-x), '--', c='#808080')
plt.plot(x, -np.exp(-x), '--', c='#505050')
plt.title(r'$f(x) = e^{-x} \cdot \cos(10x)$')
plt.savefig('myfig.pdf')
plt.show()
```
